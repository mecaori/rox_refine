```python
import random

# price 
p_ori = 2400
p_enrich_ori = 9000
p_reduce_prevent = 18700

state = [6, 6, 12, 12, 6, 6, 12, 12]

chance = [
    (100, 0 , 0 , 0), 
    (50, 50, 0, 0),
    (33, 66, 0, 0),
    (25, 75, 0, 0),
    (25, 40, 25, 10),
    (25, 40, 25, 10),
    (20, 40, 25, 15),
    (20, 40, 25, 15)
]

chance_re = [
    (100, 0 , 0 , 0), 
    (50, 50, 0, 0),
    (33, 66, 0, 0),
    (25, 75, 0, 0),
    (20, 60, 0, 20),
    (13.33, 61.67, 0, 25),
    (10, 60, 0, 30),
    (6.67, 58.33, 0, 35)
]
```

```python
def refining(refine, cost):
    # click    
    sampleList = ['success', 'fail', 'reduce', 'break']
  
    status = random.choices(
        sampleList, weights=chance[refine], k=1)
    
    # calculate cost
    if refine >= 4:
        cost = cost + state[refine]*p_enrich_ori
    else:
        cost = cost + state[refine]*p_ori
    
    # change refine value
    if status == ['success']:
        refine = refine + 1
    elif status == ['reduce']:
        refine = refine -1
    else:
        refine = refine
    
    return refine, cost
```

```python
def refining_re(refine, cost):
    # click    
    sampleList = ['success', 'fail', 'reduce', 'break']
  
    status = random.choices(
        sampleList, weights=chance_re[refine], k=1)
    
    # calculate cost
    if refine >= 4:
        cost = cost + state[refine]*p_enrich_ori + state[refine]*p_reduce_prevent
    else:
        cost = cost + state[refine]*p_ori
    
    # change refine value
    if status == ['success']:
        refine = refine + 1
    elif status == ['reduce']:
        refine = refine -1
    else:
        refine = refine
    
    return refine, cost
```

```python
# ไม่ใช้่กันลด
refine = 0
cost = 0

while refine != 8:
    refine, cost = refining(refine, cost)
    print('บวก ',refine,' ใช้เงิน ', cost)
```

```python
# ใช้กันลด
refine = 0
cost = 0

while refine != 8:
    refine, cost = refining_re(refine, cost)
    print('บวก ',refine,' ใช้เงิน ', cost)
```

```python
sum_1 = 0
sum_2 = 0

for i in range(0,10000):
    refine = 0
    cost = 0

    while refine != 8:
        refine, cost = refining(refine, cost)
    sum_1 = sum_1 + cost
    
    refine = 0
    cost = 0

    while refine != 8:
        refine, cost = refining_re(refine, cost)
    sum_2 = sum_2 + cost

if sum_1/100 < sum_2/100:
    print('ไม่ใช้กันลดดีกว่า')
else:
    print('ใช้กันลดดีกว่า')
```
